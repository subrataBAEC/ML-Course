# Question 01: [ Marks 10 ]
Explain gradient descent in detail. What is the main idea behind it? How does it help in minimizing the loss function? Explain the role of learning rate and describe what happens if the learning rate is too high or too low.

## Write Answer 01:

**Gradient descent:**

Gradient descent is an optimization algorithm to minimize the loss function by iterative process. In a NN, bias and weights are initially assigned a value, then they are updated with different itertations using the following equations,

$ W = W - \eta \frac{\partial L}{\partial W} $ .....(1)

$ b = b - \eta \frac{\partial L}{\partial b} $......(2)

here, eq 1 is for updating the weights and eq 2 is for updating the bias and $\eta$ is the learning rate.


$\frac{\partial L}{\partial b}$ is the grdient of the loss function. This represents the loss. The minus sign ensures that the updated W and b is more close to the global minimum value. This reduces the loss in the next update and performance is increased.

**Learning rate:**

Learning rate determines the amount of step with which the parameters are updated. If its values is too large, the parameter may overshoot and may loose the global minima. On the other hand, if its value is too low, updates becomes too slow and the traning becomes inefficient.

# Question 03: [ Marks 15 ]
Why do we need activation functions in neural networks? What would happen if we only used linear transformations without any activation function?

## Write Answer 03:
In ANN, output from each neuron is the weighted sum of inputs and the addition of a bias term. This is a linear output. This output is transformed into non linear through an activation function. The helps to capture the non linear relationships in the dataset.

During the forward propagation, In each neuron, a linear transformation is first applied:

$\
z = W x + b \$

here, W is the weight vector, x is the input vector and b is the bias term.


Then an activation function is applied:
$\
a = f(z)\ $

Without activation function, it only produce linear output no matter how many layers are used.


If activation function is not used, the ANN can not identify the non linear relationship. Without activation function, ANN can not creat complex decision boundaries which is essential for real world data.

# Question 04: [ Marks 15 ]

Explain the sigmoid activation function. Where is it used? What are its advantages and disadvantages? What problem does it cause during training?

## Write Answer 04:
The sigmoid activation function is a non linear activation function that gives outputs between 0 and 1. The equation is given below.

$$
\sigma(z) = \frac{1}{1 + e^{-z}}
$$

the output range is mentioned below

$$\
0 < \sigma(z) < 1
\$$

Sigmoid activation is primarily used in binary classification. it gives output in probabilities between 0 and 1.

Advantages:
1. outputs are in probabilities.

2. the derivatrive of the sigmoid function is given below.

$$\
\sigma'(z) = \sigma(z)\big(1 - \sigma(z)\big)
\$$
is is easy to compute because the derivative can be computed from the output itself. there are no need for extra calculation.

Disadvantages:
1. faces vanishing gradient problem. gradient becomes very low for very large or small values.
2. output is always positive it can cause slowing down of training.
3. for large values learning is slow compared to other activation functions.


Problems during traning:
the main problem is the vanishing gradient issue.
The derivative explains why sigmoid can cause problems.

When z is very large, σ(z)≈1

and when z is very small σ(z)≈0

but in both cases, $\
\sigma'(z) = \sigma(z)\big(1 - \sigma(z)\big)=0
\$, gradient becomes very small. this slows the learning.

# Question 05: [ Marks 10 ]
Explain the softmax activation function in simple words. Why is it used in multi-class classification problems? How is it different from sigmoid?

## Write Answer 05:
Softmax activation function is used to convert multiple inputs into probability distribution across multiple classes.The outputs are between 0 and 1 and the sum of all the outputs is 1.

$$
\text{softmax}(z_i) = \frac{e^{z_i}}{\sum_{j=1}^{K} e^{z_j}}
$$

here, i denotes a specific class under consideration, and j is used to iterate over all classes.

for example, if outputs from a softmax activation functions are such that,
dog=0.7, cat=0.2 and cow=0.1, the output is dog.


Reasons behind multiclass classifiction:
1. it produces probability for each class and sum of all probabilities equal to 1.
2. The class with the highest probabilty is the final prediction or output.

Difference with sigmoid:
1. the main difference is, softmax is more ideal for multiclass classification problems than sigmoid.
2. sum or outputs are always 1.
3. the output of a class depends on other classes too, because sum of all probabilities is 1.
