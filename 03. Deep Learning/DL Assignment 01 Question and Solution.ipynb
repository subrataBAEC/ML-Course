# Question 01: [ Marks 10 ]

What is a perceptron?
Explain its three main components and their roles.

## Write Answer 01:

A perceptron is the simplest type of neural network model which can solve simple classifications such as yes or no. It has 3 components basically-

a. inputs: These are the inputs that are provided into the perceptron. such as study hours, number of friends, age etc. It can be denoted by x1,x2,x3 etc. its role is to provide information to the perceptron.

b. weights: each Input has weights depending on the importance of that input. can be denoted by w1, w2 etc. it expresses the importance of an input feature.

c. bias: bias is a numerical term added to the equation. This helps to shift the decision boundary to make more accurate predictions.

Weights and bias are updated with each iteration to make the perfect decision boundary.

The perceptron is called linear classifier because it classifies binary problems using a straigt line or a plane with higher inputs. also it can only solve those problems where the data are linearly separable.

# Question 04: [ Marks 15 ]

Write the perceptron weight update rule and explain it intuitively.
Why are the weights updated only when the prediction is incorrect?

## Write Answer 04:
the perceptron weight update rule is:

w_new = w + η * (t - y) * x

where,

w_new= new weight, w = present weight, η = learning rate, t = target (true output), y = predicted output, x = input value

from this equation it is clear that, If the prediction is wrong, it changes the weight to make the next prediction better.If the prediction is correct, no change is needed.

the weight updates only when (t - y) is non zero. on the other hand, when the prediction is correct it is zero. hence, no change is needed for the weight. also a small leanring rate is considered so that the change is very small.

# Question 06: [ Marks 20 ]

What is meant by the “perceptron Learning Rule” or weight adjustment method?
Why can adjusting one weight affect the classification of other points?

## Write Answer 06:
The perceptron learning rule means that, the weights are updated using the following equation,

w_new = w + η * (t - y) * x

so, if the prediction is correct there is no change in the weight. weight updates only when the prediction is incorrect.
 there are two cases possible, when the output (y) is low (y<t) the weight will be increased and if the output (y) is high (y>t) that the weight will be decreased.

 adjusting one weight can affect other  points because it changes the decision boundary. changing weights changes the decision boundary. each point is considered one at a time whether the prediction is correct or not. so, an output in the correct region may fall in the wrong region due to change in weights or decision boundary for the other points.
